## III. Data Cleaning

```python
# Check if there's any duplicated value
df.duplicated().any() # False : No duplicated value
```

```python
# Check how many rows which contains null value
df.isnull().sum()
```

```python
# Replace null values in each column 2015-2020 with median value
# Median value is suitable because there are a lot of outliers value in a column, that we should not use mean()
for x in ['2015','2016','2017','2018','2019','2020']:
    df[x].fillna(df[x].median(),inplace = True)
df
```

```python
# Check if there's any null value left
df.isnull().any() # False : No null value
```

## VI. Exploratory Data Analysis

1) In which year are there most visitors coming to Asean Countries? During the Peak year and Low year, which asean countries have most vistors?

```python
# Calculate total of visitors by years
total_visitors_year = pd.Series([df['2015'].sum(),
                               df['2016'].sum(),
                               df['2017'].sum(),
                               df['2018'].sum(),
                               df['2019'].sum(),
                               df['2020'].sum()],index=['2015','2016','2017','2018','2019','2020'])
print(total_visitors_year)
```

```python
year = ['2015','2016','2017','2018','2019','2020']
```

```python
plt.plot(year,total_visitors_year,marker = 'o',mfc = 'r')
plt.title("Total of Visitors to Asean Countries by Years")
plt.xlabel("Years")
plt.ylabel("Numbers of Visitor")
plt.show()
```

@@ Result Analysis : About line plot chart, it shows that numbers of visitors increased from above 100 millions in 2015 to 140 milllions in 2019. However, this trends fell down considerably in 2020 , just about 25 millions in 2020.      

```python
# Total of visitors to Asean countries during 2016
v_asean_2019 = pd.Series(df.groupby('Destination Country')['2019'].sum())
v_asean_2019
```

```python
plt.pie(v_asean_2019, labels= v_asean_2019.index,autopct='%1.1f%%')
plt.title("Numbers of Visitors to Asean Countries in 2019(Peak Time)")
plt.show()
```

@@ Result Analysis : In Above line chart showing 2019 is the Peak time when having most visitors to Asean Countries. Among those Asean countries, Thailand,Malaysia,Singapores,Vietnam, and Indonesia are favorable countries which attracts most visitors. In addition, Cambodia had 4.7% of visitors from around the world coming to visit, which was noticeable high too.

Numbers of visitors to each country of Asean during 2020(Low Year)

```python
# Total of visitors to Asean countries during 2020
v_asean_2020 = pd.Series(df.groupby('Destination Country')['2020'].sum())
v_asean_2020
```

```python
plt.pie(v_asean_2020, labels= v_asean_2020.index,autopct='%1.1f%%')
plt.title("Numbers of Visitors to Asean Countries in 2020(Low Time)")
plt.show()
```

@@ Result Analysis : Although, during 2020 (Low time) the numbers of visitors to Asean countries decrease remarkably, we have seen Thailand,Malaysia,Singapores,Vietnam, and Indonesia still keep being the most favorable countries, and Vietnam surpassed Singapore in numbers of visitors.

2) Show total numbers of visitors coming to each countries in Asean , and which are the top 5 countries having most visitors?

```python
# Total of visitors to each asean country year by year , v_ec: visitors to each country in Asean
v_ec_15 = pd.Series(df.groupby('Destination Country')['2015'].sum())
v_ec_16 = pd.Series(df.groupby('Destination Country')['2016'].sum())
v_ec_17 = pd.Series(df.groupby('Destination Country')['2017'].sum())
v_ec_18 = pd.Series(df.groupby('Destination Country')['2018'].sum())
v_ec_19 = pd.Series(df.groupby('Destination Country')['2019'].sum())
v_ec_20 = pd.Series(df.groupby('Destination Country')['2020'].sum())
```

```python
# v_ec is Dataset showing total numbers of visitor to each Asean country in each year
v_ec = pd.DataFrame([v_ec_15,v_ec_16,v_ec_17,v_ec_18,v_ec_19,v_ec_20])
v_ec
```

```python
# Using group-bar chart to show each asean countries' visitors by year
v_ec_plot = v_ec.plot(kind= 'bar', stacked = False, title='Number of Visitors to Each Asean Country by Year',figsize =(10,5))
v_ec_plot.legend(title = 'Asean Countries',loc='center left', bbox_to_anchor=(1, 0.5))
```

@@ Result Analysis : From 2015-2020, Thailand,Malaysia,Singapores,Vietnam, and Indonesia are top 5 countries having most visitors among Asean countries, while Brunei Darussalam had the least visitors. Cambodia was top 7th having most visiors

3) Show average numbers of visitors coming to each countries in Asean per year

```python
# Average numbers of visitors per year in each country in Asean during 2015-2020
v_ec_avg = v_ec.mean().astype('i')
v_ec_avg
```

```python
v_ec_avg.plot(kind='bar',xlabel='Asean Countries',ylabel='Numbers of Visitors',title='Average Numbers of Visitors per year in Each Country of Asean During 2015-2020',color='g')
```

4) Which are the top 10 countries in the World having most visitors coming to Asean countries?

```python
# Total number of visitors from each origin countries coming to Asean Countries by year

org_15 = pd.Series(df.groupby('Origin Country')['2015'].sum())
org_16 = pd.Series(df.groupby('Origin Country')['2016'].sum())
org_17 = pd.Series(df.groupby('Origin Country')['2017'].sum())
org_18 = pd.Series(df.groupby('Origin Country')['2018'].sum())
org_19 = pd.Series(df.groupby('Origin Country')['2019'].sum())
org_20 = pd.Series(df.groupby('Origin Country')['2020'].sum())
# org is Dateset of numbers of visitors from each origin countries

org = pd.DataFrame([org_15,org_16,org_17,org_18,org_19,org_20])

# Find mean value from number of visitors in whole duration 2015-2020, and take only top 10 highest mean value

avg_org = org.mean().nlargest(10)
avg_org
```

```python
# Rank top 10 countries by using horizontal bar chart
avg_org.sort_values().plot(kind='barh',xlabel='Numbers of Visitors',ylabel='Countries of Visitors',title='Top 10 Countries Having Most Visitors to Asean Countries')
```

@@ Result Analysis: Above horizontal bar chart showed countries from Top 1 (China) to Top 10 (India) which are countries having most people visiting asean countries . We could see some countries in Asean including Singapore,Malaysia,Indonesia, and Thailand also have great numbers of people going to visit other Asean Countries.

5) Which are the top 10 countries in the World having most visitors coming to Cambodia? 

```python
# From df Datafram, Select dataframe only for 'Destination Country': 'Cambodia'
v_cam = df.loc[df['Destination Country']=='Cambodia']
v_cam
```

```python
# Find mean values for visitors during 2015-2020
v_cam.loc[:,'Average Visitors'] = pd.Series(v_cam.mean(axis=1,numeric_only=True).astype('i'))
v_cam
```

```python
top10_v_cam = v_cam[['Origin Country','Average Visitors']].nlargest(n=10,columns='Average Visitors')
top10_v_cam
```

```python
top10_v_cam.sort_values('Average Visitors').plot.barh('Origin Country',xlabel='Numbers of Visitors',ylabel='Countries of Visitors',title='Top 10 Countries having most visitors to Cambodia',color='r')
```

@@ Result Analysis: Above horizontal bar chart showed China was the country with the most people who came to visit Cambodia among other nations following by Vietnam and Thailand .
